# **AlexNet**

Introduced by Krizhevsky et al. in [ImageNet Classification with Deep Convolutional Neural Networks](https://proceedings.neurips.cc/paper/2012/file/c399862d3b9d6b76c8436e924a68c45b-Paper.pdf). AlexNet is a classic convolutional neural network architecture. It consists of convolutions, max pooling and dense layers as the basic building blocks. Grouped convolutions are used in order to fit the model across two GPUs.

## Abstract

This model explores the AlexNet model architecture proposed in [ImageNet Classification with Deep Convolutional Neural Networks](https://proceedings.neurips.cc/paper/2012/file/c399862d3b9d6b76c8436e924a68c45b-Paper.pdf), as well as its implementation with Keras.

## AlexNet Architecture

The architecture of AlexNet contains eight learned layers - five convolutional and three fully-connected. 

From the architecture shown below, the following observations can be made:

- The first convolution reduces the image dimensions considerably (Stem Conv).
- As the network goes deeper, the image dimensions are reduced further.
- As the network goes deeper, the number of kernals increases (except for the last conv layer due to the computational cost).
- Most of the parameters of the network are of the FC layers.

## AlexNet Implementation with Keras

### Imports

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPool2D
```

### AlexNet Model Creation

```python
alexnet_model = Sequential([
    
    # Layer 1    
    Conv2D(filters=96, kernel_size=(11, 11), strides=(4, 4), activation="relu", input_shape=(227, 227, 3)),
    MaxPool2D(pool_size=(3, 3), strides=(2, 2)),

    # Layer 2
    Conv2D(filters=256, kernel_size=(5, 5), strides=(1, 1), activation="relu", padding="same"),
    MaxPool2D(pool_size=(3, 3), strides=(2, 2)),
  
    # Layer 3, 4, 5
    Conv2D(filters=384, kernel_size=(3, 3), strides=(1, 1), activation="relu", padding="same"),
    Conv2D(filters=384, kernel_size=(3, 3), strides=(1, 1), activation="relu", padding="same"),
    Conv2D(filters=256, kernel_size=(3, 3), strides=(1, 1), activation="relu", padding="same"),
    MaxPool2D(pool_size=(3, 3), strides=(2, 2)),
    
    # Layer 6, 7, 8
    Flatten(),
    Dense(4096, activation="relu"),
    Dense(4096, activation="relu"),
    Dense(10, activation="softmax")

])

alexnet_model.summary()
```
