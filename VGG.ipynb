## Introduction

The VGG model was introduced to improve the accuracy of the original ConvNet architecture published in [ImageNet Classification with Deep Convolutional
Neural Networks](https://proceedings.neurips.cc/paper/2012/file/c399862d3b9d6b76c8436e924a68c45b-Paper.pdf). It does so by fixing other parameters of the architecture, and steadily increasing the depth of the network by adding more convolutional layers, which is feasible due to the use of very small (3×3) convolution filters in all layers.

## Model Architecture

Three main trends are enforced in the VGG model:
- (3x3) convolutions with $S=1$, $P=1$.
- (2x2) max pooling with $S=2$, $P=0$.
- Doubling of the number of channels after each pooling operation.

## VGG Limitations

- One issue with the VGG model is the lack of stem layers at the beginning of the network. This significantly increases the amount of memory required to store all the model's activations.

- Another issue that is in common with AlexNet is that the majority of the network parameters is condensed in the last fully connected layers.
